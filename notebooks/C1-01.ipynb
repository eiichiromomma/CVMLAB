# Symbolicなプロット

```python
from sympy import symbols, sin, pi
from sympy.plotting import plot
t = symbols('t')
v = 141 * sin(pi*t+pi/4)
display(v)
```

```python
plot(v, (t, -2,2));
```

SymPyもMATLAB同様に微分方程式が扱える(=0の形にする必要がある)

```python
from sympy import Function, dsolve, Derivative
from sympy.abc import x
y = Function('y')
eq = Derivative(y(x),x) - 2*(x+1)
eq
```

```python
dsolve(eq,y(x))
```

```python
dsolve(eq, y(x), ics={y(x).subs(x,0):1})
```
